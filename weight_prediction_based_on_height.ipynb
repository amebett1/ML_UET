```python
import numpy as np 
import matplotlib.pyplot as plt
```

Data Visualization

```python
# Height (cm)
X = np.array([147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]).reshape(-1, 1)
# Weight (kg)
y = np.array([49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]).reshape(-1, 1)
print('Chiều cao (cm): ', X)
print('Cân nặng (kg):', y)
```

```python
plt.scatter(X,y)
plt.axis([140,200,40,80])
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Cân nặng (kg)')
```

Tìm Fitting Line

```python
new_column_2d = np.ones((X.shape[0], 1))
Xbar = np.concatenate((new_column_2d, X), axis=1)

# 3. Tính A và b dựa trên Xbar (để bao gồm hệ số chặn)
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y)

print("--- Ma trận Xbar ---")
print(Xbar)
print("\n--- Ma trận A (Xbar.T * Xbar) ---")
print(A)
print("\n--- Vector b (Xbar.T * y) ---")
print(b)

w = np.dot(np.linalg.pinv(A),b) #A dagger * b
print('w = ', w)
```
