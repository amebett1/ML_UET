```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([
    30, 32.41, 34.83, 37.24, 39.66, 40.34, 42.41, 44.48, 46.55, 48.62,
    50.69, 52.76, 54.83, 56.9, 58.97, 61.03, 63.1, 65.17, 67.24, 69.31,
    71.38, 73.45, 75.52, 77.59, 79.66, 81.72, 83.79, 85.86, 87.93, 90
]).reshape(-1,1)
y = np.array([
    448.524, 509.248, 535.104, 551.432, 623.418, 618.456, 688.32, 705.682, 
    714.541, 768.38, 781.857, 815.409, 863.16, 853.653, 891.017, 947.868, 
    972.454, 1032.593, 1041.745, 1065.257, 1156.414, 1156.183, 1195.646, 
    1199.398, 1250.603, 1297.307, 1305.666, 1369.797, 1383.868, 1423.645
]).reshape(-1,1)
```

```python
plt.xlabel('Area (m2)')
plt.ylabel('Price (Million VND)')
plt.title('Data')
plt.grid(True)
plt.scatter(X,y)
plt.plot()
```

$$\mathbf{w} = (\bar{\mathbf{X}}^T \bar{\mathbf{X}})^{-1} \bar{\mathbf{X}}^T \mathbf{y}$$

```python
new_column_2d = np.ones((X.shape[0], 1))
Xbar = np.concatenate((new_column_2d, X), axis=1)
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y)
w = np.dot(np.linalg.pinv(A),b)
print(w)
w0 = w[0][0]
w1 = w[1][0]
print(w0,w1)
```

```python
XX = np.linspace(25, 95)
yy = w1*XX + w0
plt.xlabel('Area (m2)')
plt.ylabel('Price (Million VND)')
plt.title('Data')
plt.grid(True)
plt.scatter(X,y)
plt.plot(XX,yy,color='red')
plt.show()
```
